#### Exercise 1 (Allais-Samuelson-Diamond Overlapping Generations model - setup and equilibrium)

Consider the Allais-Samuelson-Diamond OLG model with capital. An agent (born young at date $t \in \mathbb{N}$) consumes $c_{y,t}$ when young and $c_{o, t + 1}$ when old. There is an infinite set of such agents on the interval $[0, 1]$. The date-$t$ young agent has a preference ordering over $(c_{y, t}, c_{o, t + 1})$ streams which is represented by the utility function:

$$U(c_{y, t}, c_{o, t + 1}) = \ln(c_{y, t}) + \beta \ln(c_{o, t + 1})$$

Where $\beta > 0$. The sequence of budget constraints for an agent is:

$$c_{y, t} + s_t = w_t \times 1$$

and, 

$$c_{o, t + 1} = R_{t + 1}s_t$$

Where $s_t$ is savings flow, $w_t$ and $R_t - 1$, respectively, are competitive equilibrium real rental rates on labour and capital.

Capital is assumed to depreciate fully in each period. Total saving in the economy, $s_t \times 1$, equals new capital next perios: $k_{t + 1} = s_t$.

Let $k$ denote the capital-to-labour ratio, $(A, k) \rightarrow f(A, k)$ is the production function, and $(A, k) \rightarrow f_k(A, k)$ is its first-order derivative function (with respect to $k$). 

$$f(A_t, k_t) = A_t k_t^\alpha$$

The firms is perfectly competitive in both its product market and its factor input (capital and labour) market, and maximises profits.

1. Write down each generation's ex-ante decisions problem and characterise their optimal behaviour.
2. Do the same for the firm
3. Define a recursive (dymanic) competitive equilibrium (list quantitves and prices, and write down their corresponding demand/supply and market clearing conditions).

#### Answers 1.1

Each generation wishes to maximise their utility function given their budget constraint:

$$\max_{c_y, c_o} \space \ln{(c_{y, t})} + \beta \ln{(c_{o, t + 1}})$$

$$s.t.$$

$$c_{y, t} + s_t = w_t \times 1$$

$$c_{0, t + 1} = R_{t + 1} s_t$$

There are two methods to solve this problem, first by substition and then by the lagrangian method, both will be presented:
    
__Substitution__

Begin by substuting both budget constraints into the the utility function and maximising with respect to $s_t$

$$\max_{s_{t}} \space \ln{(w_t - s_t)} + \beta \ln{(R_t s_t)}$$

Since the utility function and the budget constraints are continuously differentiable, we can directly differentiate the indirect value function (using the envelope theorem) to get the first order conditions:

$$\frac{- 1}{w_t - s_t} + \beta\frac{R_t}{R_t s_t} = 0$$

$$s_t = (w_t - s_t) \beta$$

$$s_t = \frac{\beta w_t}{1 + \beta}$$

Therefore, each generations consumption behaviour is:

For consumption when young we have:

$$c_{y, t} = w_t - s_t$$

$$c_{y, t} = w_t - \frac{\beta w_t}{1 + \beta}$$

$$c_{y, t} = \frac{w_t}{1 + \beta}$$

For consumption when old we have:

$$c_{o, t + 1} = R_{t + 1}s_t$$

$$c_{o, t + 1} = \frac{\beta R_{t + 1} w_t}{1 + \beta}$$


Euler equation, is given when we substitute these two together:

$$\frac{c_{o, t + 1}}{c_{y, t}} = \beta R_{t + 1}$$

__Lagrangian__

If we write out the Lagrangian equation, we get:

$$L = \ln{(c_{y, t})} + \beta \ln{(c_{o, t + 1})} + \lambda_t(c_{y, t} + s_t - w_t) + \theta_t(c_{o, t + 1} - R_{t + 1} s_t)$$

The first order conditions are:

$$u'(c_{y, t}) - \lambda_t = 0$$

$$\beta u'(c_{o, t + 1}) - \theta_t =0$$

$$-\lambda_t + \theta_t R_{t + 1} = 0$$

This gives us the same Euler equation:

$$\beta u'(c_o, t + 1)R_{t + 1} = u'(c_{y, t})$$

#### Answer 1.2
As for the firm, we have:

$$\pi_t = F(K_t, L_t) - R_t K_t - w_t L_t$$

Where $F(K, L) = AK^{\alpha}L^{1 - \alpha}$

In equilibrium, the profits will be zero, and the first order conditions become:

$$F_K(K_t, L_t) = R_t$$

$$F_L(K_t, L_t) = w_t$$

#### Answer 1.3
A <span style="color:blue">competitive equilibrium</span> can be represented by sequences of prices $\{w_t, R_t\}$ and allocations $\{c_{y, t}, c_{o, t +1}, k_t, L_t\}$ such that:
   - Taking prices as given, the allocations $k_{t + 1}$, $c_{y, t}$ and $c_{o, t +1}$ solves the agents utility maximisation problem
   - Taken prices as given, the allocations $k_t$ and $L_t$ solve the firm's maximisation problem
   - Markets clear 

We have:

$$F(K_t,L_t) = A K_t^{\alpha} L_t^{1 - \alpha}$$

Taking the first derivative with respect to $L$

$$F_L(K_t, L_t) = A(1 - \alpha)K_t^{\alpha} L_t^{- \alpha}$$

$$ = AK_t^{\alpha} L_t^{- \alpha} - \alpha AK_t^{\alpha} L_t^{- \alpha}$$

Redefining $\frac{K}{L} = k$

$$ = Ak_t^{\alpha} - \alpha A k_t^\alpha$$

$$ = f(k_t) - k_tf'(k_t)$$

Therefore in equilibrium:

$$w_t = f(k_t) - k_tf'(k_t)$$

Now, taking the derivative with respect $K$

$$F_K(K_t, L_t) = \alpha A K^{\alpha - 1} L^{1 - \alpha}$$

$$ = \alpha A k_t^{\alpha - 1}$$

$$ = f'(k_t)$$

Therefore, in equilibrium:

$$R_t = f'(k_t)$$

Since we have that $s_t = k_{t + 1}$, substituting the equilibrium wage rate, we get the capital transition equation:

$$s_t = \frac{\beta w_t}{1 + \beta}$$

$$k_{t + 1} = \frac{\beta}{1 + \beta}\left[f(k_t) - k_t f'(k_t)\right]$$

Alternatively, it can be written:

$$k_{t + 1} = \frac{\beta (1 - \alpha) A_t k_t^{\alpha}}{1 + \beta}$$

```python
def OLG_process(T, A):
    
    # Storage space and first realisation of k
    k = np.zeros(T.size)
    k[0] = 0.01
    
    for t in range(T.size - 1):
        
        k[t + 1] = h(k[t], A[t])
    
    return k
```

#### Answer 2.2
The AR(1) process is given by:

$$\ln{(A_t)} = \rho \ln{(A_{t - 1})} + \sigma w_t$$

```python
def TFP(T):
    
    # Storage space and first realisation of A
    A = np.zeros(T.size)
    A[0] = 1.0
    
    for t in range(T.size - 1):
        
        A[t + 1] = np.exp(ρ * np.log(A[t]) + σ * np.random.randn(1))
        
    return A
```

```python
# Grid
T_min = 0.0
T_max = 100
T_size = 100
T = np.linspace(T_min, T_max, T_size)
```

```python
# Plotting Capital Accumulation
fig, ax = plt.subplots(figsize=(8, 5))
ax.set_title('Capital Accumulation')
ax.set_xlabel('$t$')
ax.set_ylabel('$k_{t + 1}$')
ax.plot(T, OLG_process(T, A = TFP(T)), lw=1, alpha=0.6)
plt.show()
```

```python
def Policy_Functions(k, A):
    
    k_next = h(k, A)
    w = (1 - α) * A * k**α
    y = A * k**α
    R = α * A * k**(α - 1)
    c_y = w - k_next
    c_o = R * k_next
    
    return k_next, w, R, c_y, c_o, y
```

```python
k = OLG_process(T, A = TFP(T))
policy = np.asarray(Policy_Functions(k, A = TFP(T)))
```

```python
# Parameters
α = 0.33
β = 1.0 / (1.0 + 0.04**35.0)
θ = 2.0
A = 1.0
```

```python
# Functions
g = lambda k_new, k, A: ((1 - α) * A * k**α) / k_new
h = lambda k_new, A: β**(- 1 / θ) * ((α * A * k_new**(α - 1))**((θ - 1) / θ)) + 1
```

```python
def Capital(k, A, k_lb=1e-20, k_ub=10.0, Dynamic=True):
    
    if Dynamic==True:
        
        f = lambda k_new, k, A: h(k_new, A) - g(k_new, k, A)
        k_new = brentq(f, k_lb, k_ub, args=(k, A))
        
    else:
        
        f = lambda k, A: h(k, A) - g(k, k, A)
        k_new = brentq(f, k_lb, k_ub, args=(A, ))
        
    return k_new
```

```python
guess_k = 1e-20

k_steady_state = Capital(guess_k, A, Dynamic=False)

k_steady_state
```

```python
# Grid
k_min = 1e-20
k_max = k_steady_state * 2
k_size = 200
k_old = np.linspace(k_min, k_max, k_size)
```

```python
# Storage space
k_new = []

for t, k, in enumerate(k_old):
    
    k_n = Capital(k, A)
    k_new.append(k_n)

k_new = np.asarray(k_new)
```

*Pseudocode (policy/time iteration method)*:

Step 0. Define state space $X \ni k_{t}$

Step 1. Define $u'$, $f$ and $f'$ functions

Step 2. Set $k_{T+1}^{\star} = 0$ for all $k_{T}$

1. Set $t \gets T$

2. Use resource constraint, solve for $c_{T}^{\star} = h_{T}(k_{T})$, for any given $k_{T}$

Step 3. 

*Input*:

* $X$, $u'$, $f$ and $f'$

(Backward induction.) For each $t = T-1, T-2, ..., 0$:

* Given $k_{t+1}^{\star}$, and,

* given function $h_{t+1}$ such that $c_{t+1}^{\star} = h_{t+1}(k_{t+1})$ from the last stage indexed by $t+1$,

* use the Euler equation, solve for $c_{t}^{\star} = h_{t}(k_{t})$ for any $k_{t}$.

*Output*: 

* Sequence of optimal consumption decision functions $\{h_{t}\}_{t=0}^{T}$.

```python
# Parameters
α = 0.33
θ = 1.0
β = 0.98
A = 1.0
```

```python
# Functions
u = lambda c: np.log(c)
f = lambda k: k**α
u_prime = lambda c: (c + 1e-18)**(- θ)
f_prime = lambda k: α * k**(α - 1)
```

```python
def Euler_Equation(w, grid):
    
    # Storage space
    k_n_opt = np.zeros(grid.size)
    c_opt = np.zeros(grid.size)
    
    # Interpolation function
    c_func = lambda x: np.interp(x, grid, w)
    
    for i, k in enumerate(grid):
        
        euler = lambda k_n, k: u_prime(f(k) - k_n) - β * u_prime(c_func(k_n)) * f_prime(k_n) 
        k_n_opt[i] = brentq(euler, 1e-12, f(k), args=(k,))
        c_opt[i] = f(k) - k_n_opt[i]
    
    return k_n_opt, c_opt

def Compute_Steady_State(grid, T):
    
    # Interpolation function
    c_func = lambda x: np.interp(x, grid, f(grid))
    
    # Storage space
    savings_policy = [np.zeros(grid.size)]
    consumption_policy = [c_func(f(grid))]
    
    # Final consumption
    w = c_func(grid)
    
    for t in range(T - 1, -1, -1):
        
        k, c = Euler_Equation(w, grid)
        savings_policy.append(k)
        consumption_policy.append(c)
        w = c
        
    savings_policy = savings_policy[::-1]
    consumption_policy = consumption_policy[::-1]
    
    return np.asarray(savings_policy), np.asarray(consumption_policy)
```

```python
# Functions
def Optimal_Path(initial_k, c, k, T):
    
    # Interpolators
    consumption = lambda x, t: np.interp(x, grid, c[t])
    savings= lambda x, t: np.interp(x, grid, k[t])
    
    # Storage space
    c_star = []
    kn_star = []
    
    # Initial k value
    kn_star.append(initial_k)
    
    for t in range(T):
        
        c_star_new = consumption(initial_k, t)
        c_star.append(c_star_new)
        
        k_star_new = savings(initial_k, t)
        kn_star.append(k_star_new)
        
        initial_k = k_star_new
        
    return c_star, kn_star
```
